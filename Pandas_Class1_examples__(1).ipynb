```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.figure(figsize=(4,4))
plt.hist(df1['Accept'])
plt.xlabel('Accept')
plt.ylabel('count')
plt.title('Histogram of Accept')
#for not displaying values
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.hist(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('Count')
plt.title('Histogram of SAT')
#for not displaying values
plt.show()
```

```python
sns.histplot(df1['Accept'])
```

```python
sns.histplot(df1['SAT'])
```

```python
sns.distplot(df1['Accept'])
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.displot(df1['Accept'])
```

```python
plt.plot(df1['Accept'],df1['Expenses'],'r+')
plt.xlabel('Accept')
plt.ylabel('Expenses')
plt.title('Scatterplot of Accept vs Expenses')
```

```python
#want to know any of the function
help(plt.plot)
help(plt.hist)
```
